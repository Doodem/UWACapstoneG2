```python
from docx.api import Document
import os
import re
import nltk
from nltk import word_tokenize

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer

import spacy
```

```python
def get_all_text_list(a):
    text = []
    for item in a.paragraphs:
        if item.text != "":
            text.append(item.text)
    return text
```

```python
def get_all_headings(document):
    Headings = []
    for item in document.paragraphs:
        if item.style.name.startswith("Heading"):
            Headings.append(item.text)
    return Headings
```

```python
nltk.download('punkt')
def pre_process(text):
    processed_text = []
    for item in text:
        item = item.lower()
        for word, new_word in contraction_dict.items():
            item = item.replace(word, new_word)
        item = re.sub(r'[^\w\s]',' ',item)
        output = word_tokenize(item)
        processed_text.extend(output)
    return processed_text
```

```python
def get_tfidf(docs):
    tfidf_dict = dict()
    vectorizer = TfidfVectorizer(stop_words="english")
    tfidf_matrix = vectorizer.fit_transform([docs])
    tfidf_array = tfidf_matrix.toarray()
    terms = vectorizer.get_feature_names_out()
    for x,y in zip(terms,tfidf_array[0]):
        tfidf_dict[x] = y
    return tfidf_dict
```

```python
def collapse_list_of_strings(x):
    l = ""
    for item in x:
        l = l + " " + item
    return l 
```
